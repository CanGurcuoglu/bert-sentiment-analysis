```python
import pandas as pd
from googletrans import Translator
import ssl

# CSV dosyasını okuma
df = pd.read_csv('data1.csv')

# Çevirici nesnesi oluşturma
ssl._create_default_https_context = ssl._create_unverified_context
translator = Translator()

# Çeviri ve uyarlama fonksiyonu
def translate_and_adapt(comment):
    # İngilizce'den Türkçe'ye çeviri
    translated = translator.translate(comment, src='en', dest='tr').text
    # Şirket adını uyarlama
    adapted = translated.replace('Vodafone', 'Türk Telekom')
    return adapted

# Yeni bir sütun ekleyerek çeviriyi uygulama
df['yorum_tr'] = df['review'].apply(translate_and_adapt)
```
