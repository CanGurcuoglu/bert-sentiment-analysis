```python
import os

def dosya_parcala(dosya_adi, parca_sayisi):
    """
    Belirtilen dosyayı belirtilen sayıda parçaya böler.

    Args:
        dosya_adi: Parçalanacak dosyanın adı.
        parca_sayisi: Dosyanın bölüneceği parça sayısı.
    """
    parca_boyutu = os.path.getsize(dosya_adi) // parca_sayisi
    with open(dosya_adi, 'rb') as f:
        for i in range(parca_sayisi):
            parca_adi = f'{dosya_adi}-{i+1}'
            with open(parca_adi, 'wb') as parca_dosyasi:
                parca_dosyasi.write(f.read(parca_boyutu))

def dosya_birlestir(dosya_adi, parca_sayisi):
    """
    Belirtilen dosyanın parçalarını birleştirerek orijinal dosyayı oluşturur.

    Args:
        dosya_adi: Birleştirilecek dosyanın adı.
        parca_sayisi: Dosyanın parça sayısı.
    """
    with open(dosya_adi, 'wb') as cikti_dosyasi:
        for i in range(parca_sayisi):
            parca_adi = f'{dosya_adi}-{i+1}'
            with open(parca_adi, 'rb') as parca_dosyasi:
                cikti_dosyasi.write(parca_dosyasi.read())

```

```python
# Örnek kullanım:
dosya_parcala('model.safetensors', 50)  # data.txt dosyasını 10 parçaya böler.
```

```python
dosya_birlestir('model.safetensors', 10)  # data.txt dosyasını parçalarından birleştirir.
```
